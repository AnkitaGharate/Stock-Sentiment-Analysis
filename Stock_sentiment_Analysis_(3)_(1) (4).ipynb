```python
import pandas as pd
pd.set_option("display.max_columns",None)
```

```python
# Removing punctuations
data=df.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)
```

```python
data
```

```python
# Renaming column names for ease of access
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(5)
```

```python
# Convertng headlines to lower case
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(2)
```

```python
import nltk
nltk.download('stopwords')
stopwords=nltk.corpus.stopwords.words('english')
```

```python
data.head
```

```python
' '.join(str(x) for x in data.iloc[0,0:25])
```

```python
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
data["headlines"] = pd.Series(headlines)
```

```python
data["headlines"][0]
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download("punkt")
```

```python
def token(text):
  word_tokn = word_tokenize(text,language = "english")
  return word_tokn
```

```python
data["headlines"] = data["headlines"].apply(token)
```

```python
def remove_stopwords(text):
  txt_clean=[word for word in text if word not in stopwords]
  return txt_clean

data["headlines"] = data["headlines"].apply(remove_stopwords)
```

```python
# data["headlines"][0]
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

def stemin(text):
  words = [ps.stem(i) for i in text]
  return words
```

```python
data["headlines"] = data["headlines"].apply(stemin)
```

```python
# data["headlines"][0]
```

```python
def joint(text):
  headline = [i for i in text]
  return ' '.join(headline)
```

```python
data["headlines"] = data["headlines"].apply(joint)
```

```python
from sklearn.model_selection import train_test_split
X = data["headlines"]
y = df["Label"]
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 42,test_size = 0.2)
```

```python
from sklearn import preprocessing
```

```python
from sklearn.feature_extraction.text import CountVectorizer
countvector=CountVectorizer()
X_train_tfm=countvector.fit_transform(X_train)
X_test_tfm=countvector.transform(X_test)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidftvector=TfidfVectorizer(ngram_range=(2,2))
X_train_tfm1=tfidftvector.fit_transform(X_train)
X_test_tfm1=tfidftvector.transform(X_test)
```

```python
from imblearn.combine import SMOTETomek
smt=SMOTETomek(random_state=42)
X_train_tfm1,y_train=smt.fit_resample(X_train_tfm1,y_train)
```

```python
y_train.value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='gini')
randomclassifier.fit(X_train_tfm1,y_train)
```

```python
rf_y_pred = randomclassifier.predict(X_test_tfm1)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(classification_report(y_test,rf_y_pred))
```
